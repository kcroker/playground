```python
#!/usr/bin/python3

from scipy.integrate import odeint
import sys
import matplotlib.pyplot as plt
import numpy as np

# This is your equation
def thesystem(state, t):
    global m
    # u := da/dt
    a, u = state

    dadt = u
    dudt = m*(a - 1/a)

    return [dadt, dudt]
```

```python
# Joel, adjust these to what you want
m = 0.2
a_init = 0.4 
dadt_init = 1 

# Make a grid to solve on
tlist = np.linspace(0, 1, 300)

# Solve
evolution = odeint(thesystem,
                   [a_init, dadt_init],
                   tlist)

# Plot
plt.plot(tlist, evolution[:,0], label='a(t)')
plt.plot(tlist, evolution[:,1], label='da/dt(t)')
plt.legend(loc='upper left')
plt.show()
```
